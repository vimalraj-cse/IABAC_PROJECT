### 2. Top 3 Important Factors Affecting Employee Performance

```python
# Prepare data for feature importance analysis
# Convert categorical variables to numerical
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
categorical_cols = ['Gender', 'EducationBackground', 'MaritalStatus', 'EmpDepartment', 
                   'EmpJobRole', 'BusinessTravelFrequency', 'OverTime', 'Attrition']

for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Feature selection
X = data.drop([ 'PerformanceRating', 'Attrition'], axis=1)
y = data['PerformanceRating']

# Train a Random Forest model to get feature importance
from sklearn.ensemble import *
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Get feature importances
feature_importances = pd.DataFrame({'feature': X.columns, 'importance': model.feature_importances_})
feature_importances = feature_importances.sort_values('importance', ascending=False)

print("\nTop 10 Important Factors Affecting Performance:")
print(feature_importances.head(10))

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='importance', y='feature', data=feature_importances.head(10))
plt.title('Top 10 Factors Affecting Employee Performance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

### 3. Trained Model to Predict Employee Performance

</b>Note</b></n>

This Trained Model to Predict Employee Performance you can be found in Train model and prediction model ipynb file

```python
# Creating a dictionary for models and their f1 scores
comparison_dict = {'Model': ['LogisticRegression','DecisionTreeClassifier','RandomForestClassifier','KNeighborsClassifier','MLPClassifier','SVC'],
                   'accuracy_score': [ 0.752778,0.600000,0.772222, 0.525000,0.841667,0.838889]}
# Creating DataFrame
comparison_df = pd.DataFrame(comparison_dict)
print(comparison_df.to_string(index=False))
```

```python
# Visualization
comparison_df = comparison_df.sort_values(by='accuracy_score', ascending=True)

max_index = comparison_df['accuracy_score'].idxmax()  # This returns the row index having highest accuracy_score

plt.figure(figsize=(16, 5))
sns.lineplot(data=comparison_df, x='Model', y='accuracy_score', marker='o', color='blue', label="Model Performance")

plt.scatter(comparison_df.loc[max_index, 'Model'], comparison_df.loc[max_index, 'accuracy_score'],color='red', s=100, label="Best Model")

plt.xlabel("Model")
plt.ylabel("accuracy_score")
plt.legend()
plt.title("Comparison of Model Performance (accuracy_score)")
plt.show()
```

#### Insights

1.Best Model: MLPClassifier (84.17%), followed by SVC (83.89%)—both perform well for complex patterns.

2.RandomForest (77.22%) shows good results, benefiting from ensemble learning.

3.Logistic Regression (75.28%) is decent and interpretable.

4.Weakest Models: KNN (52.50%) & Decision Tree (60.00%), likely due to overfitting or sensitivity to noise.

### 4. Recommendations to Improve Employee Performance

### Based on the analysis, here are actionable recommendations:

#### Compensation Strategy

Implement a more performance-based compensation structure since salary hikes showed the strongest correlation with performance

Consider more frequent, smaller salary adjustments rather than infrequent large hikes

#### Career Development 

Create clear career progression paths as job level was a key performance factor

Implement more frequent promotions for high performers to maintain motivation

Provide leadership training for employees at higher job levels

#### Experience Utilization 

Better leverage experienced employees by creating mentorship programs

Pair junior employees with senior mentors to transfer knowledge and skills

Consider experience when assigning critical projects

#### Department-Specific Improvements 

For Human Resources (lowest performing department):

Conduct engagement surveys to identify specific issues

Review workload and job design in HR department

For Data Science (highest performing department):

Identify and replicate best practices from this department

Consider rotating high-potential employees through this department

#### Training & Development 

Focus training on employees with mid-range experience (5-10 years) who showed performance dips

Implement more frequent, targeted training sessions rather than annual training events

#### Work-Life Balance 

Address work-life balance concerns as this was a moderately important factor

Implement flexible work arrangements where possible

Monitor overtime hours and their impact on performance

#### Hiring Strategy 

Use the predictive model to evaluate candidates during hiring

Focus on candidates with characteristics that predict higher performance

Consider experience level and past career progression when hiring

#### Performance Monitoring

Implement early warning systems to identify performance drops

Use the model to flag employees at risk of performance decline

Intervene early with support and coaching
