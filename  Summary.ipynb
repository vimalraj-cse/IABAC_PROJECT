# **Project Summary**  
## **INX Future Inc Employee Performance - Project**  

The **Data Science Project** presented here focuses on analyzing employee performance. The primary goal is to predict employee performance ratings based on various features such as:  
- `TotalWorkExperienceInYears`  
- `EmpDepartment`  
- `Gender`  
- `ExperienceYearsInCurrentRole`, among others.

---

### **Goals and Insights**
1. **Department-Wise Performance Analysis**  
   Understand how performance varies across different departments.  
   
2. **Top 3 Important Factors Influencing Employee Performance**  
   Identify key factors that significantly impact employee performance.  
   
3. **Performance Prediction Model**  
   Develop a machine learning model to predict employee performance based on input factors. This will aid in hiring decisions.  

4. **Actionable Recommendations**  
   Provide recommendations to improve employee performance based on analytical insights.

---

### **Dataset Overview**
- **Number of Records:** 1200 rows  
- **Number of Features:** 28 columns  
- **Data Composition:**  
  - **Quantitative Features:** 19 columns  
    - 11 numeric  
    - 8 ordinal  
  - **Qualitative Features:** 8 columns  
  - **EmpNumber:** Alphanumeric data (excluded as it doesn’t contribute to performance rating).  
- **Dataset Shape:** 1200 × 28  

---

### **Key Methodologies**
1. **Correlation Analysis**  
   - Evaluate relationships between features and performance rating using correlation, a statistical measure of linear dependency.

2. **Machine Learning Model**  
   - Model Used: **MLPClassifier**  
   - Reason: Handles categorical and numerical data effectively, achieving a **84% accuracy**.  
   - Feature Importance: Identify the most critical factors affecting performance through the model's feature importance technique.

3. **Data Preprocessing**  
   - **Label Encoding:** Convert string categorical data into numerical format for compatibility with machine learning algorithms.  

4. **Visualization Techniques**  
   - Visual insights were leveraged to uncover patterns and trends in the data.

---

### **Achievements**
- Successfully identified department-wise performance trends.  
- Determined the top 10 important factors influencing performance.  
- Built a robust predictive model with high accuracy to aid hiring decisions.  
- Offered actionable recommendations to improve employee performance based on insights.
