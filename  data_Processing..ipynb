```python
# Defining a function for checking  skewness
def skewedness(data, num_column):
    skewed = []
    for col in num_column:
        sk = data[col].skew()
        if sk >= 1.96 or sk <= -1.96:
            skewed.append(col)
    return skewed

# Print skewed columns
print(f"Skewed Column Names: {skewedness(data, num_column)}")
print("Range of skewness & kurtosis: S < |1.96|")
```

## HANDLING OUTLIERS

```python
for i in num_column:
    sns.boxplot(data,x=i)
    plt.show()
```

```python
## Outlier handling for NumCompaniesWorked
sns.boxplot(data,x='NumCompaniesWorked')
plt.show()
```

```python
#Identifying the outliers using IQR method
#finding Q1 and Q3
Q1=np.quantile(data['NumCompaniesWorked'],0.25)
Q3=np.quantile(data['NumCompaniesWorked'],0.75)
#finding IQR
IQR=Q3-Q1
```

```python
#finding lower_limit and upper_limit
lower_limit=Q1-(1.5*IQR)
upper_limit=Q3+(1.5*IQR) 
print(upper_limit)
```

```python
#Extracting the outliers from data
data.loc[(data.NumCompaniesWorked<lower_limit) | (data.NumCompaniesWorked>upper_limit)]
```

```python
# Apply capping
data['NumCompaniesWorked'] = np.where(data['NumCompaniesWorked'] < lower_limit, lower_limit, data['NumCompaniesWorked'])
data['NumCompaniesWorked'] = np.where(data['NumCompaniesWorked'] > upper_limit, upper_limit, data['NumCompaniesWorked'])
```

```python
##checking  Outlier  for NumCompaniesWorked after handled
sns.boxplot(data,x='NumCompaniesWorked')
plt.show()
```
